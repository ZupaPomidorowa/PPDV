```python
import numpy as np
import ipywidgets as w
import matplotlib.pyplot as plt
```

```python
from typing_extensions import ParamSpecKwargs
sA1 = w.FloatSlider(min=0, max=10, step=1, description='A_1', value=1.0)
lblA1 = w.Label('[V]')
sA2 = w.FloatSlider(min=0, max=10, step=1, description='A_2', value=1.0)
lblA2 = w.Label('[V]')

sPhi1 = w.FloatSlider(min=-360, max=360, step=10, description='Phi_1', value=0.0)
lblPhi1 = w.Label('[Degrees]')
sPhi2 = w.FloatSlider(min=-360, max=360, step=10, description='Phi_2', value=0.0)
lblPhi2 = w.Label('[Degrees]')

txtF1 = w.Text(description='f_1', value='1.0')
lblF1 = w.Label('[kHz]')
txtF2 = w.Text(description='f_2', value='1.0')
lblF2 = w.Label('[kHz]')

sW1 = w.FloatSlider(min=0, max=100, step=5, description='w_1', value=1.0)
lblW1 = w.Label('[Rad/s]')
sW2 = w.FloatSlider(min=0, max=100, step=5, description='w_2', value=1.0)
lblW2 = w.Label('[Rad/s]')


cbo = w.Dropdown(
    options = ['frequency', 'angular velocity'],
    description = 'Use:',
    #ensure_option = True,
    disabled=False
)

gb = w.GridBox(layout=w.Layout(grid_template_columns='320px 100px'))
gb.children = [
    sA1, lblA1,
    sA2, lblA2,
    sPhi1, lblPhi1,
    sPhi2, lblPhi2,
    txtF1, lblF1,
    txtF2, lblF2,
    sW1, lblW1,
    sW2, lblW2,
    cbo
]

out = w.Output()
x = np.arange(0, 2.0, 0.001)

def on_data_changed_callback(_):
  with out:
    #out.clear_output(True)
    w1 = None
    w2 = None
    try:
      if cbo.value == 'frequency':
        w1 = 2*np.pi*float(txtF1.value)
        w2 = 2*np.pi*float(txtF2.value)
      else:
        w1=sW1.value
        w2=sW2.value
    except:
      pass

    if w1 is not None and w2 is not None:
      plt.plot(x, sA1.value * np.sin(w1*x - sPhi1.value/360 * 2 * np.pi))
      plt.plot(x, sA2.value * np.cos(w2*x - sPhi2.value/360 * 2 * np.pi))
      A = np.max((sA1.value, sA2.value))
      plt.ylim([-A,A])
      plt.show()

for c in gb.children:
  c.observe(on_data_changed_callback)

#sA1.observe(on_data_changed_callback)
#sA2.observe(on_data_changed_callback)
#sPhi1.observe(on_data_changed_callback)
#sPhi2.observe(on_data_changed_callback)
#txtF1.observe(on_data_changed_callback)
#txtF2.observe(on_data_changed_callback)
#sW1.observe(on_data_changed_callback)
#sW2.observe(on_data_changed_callback)
#cbo.observe(on_data_changed_callback)


on_data_changed_callback(None)

w.VBox(children=[gb, out])
```
