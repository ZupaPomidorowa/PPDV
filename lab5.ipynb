Exercise 2

Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

for ax in (ax1, ax2):
  ax.plot(x,y, color='#FF0000')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
```

Exercise 4

Use plt.subplots(nrows=1, ncols=2) to create the plot below.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'b--', lw=3)
axes[1].plot(x, z, 'r', lw=3)
fig.tight_layout()
```

Using artist layer create a function which will draw an UML actor 'stickman'. The function should take three arguments axes and x,y coordinates. It should follow the following template

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.path import Path
from matplotlib.patches import Circle, PathPatch
import matplotlib.patches as patches

def actor(ax,x,y):
    points = [
        [0.2, 0.2],
        [0.3, 0.4],
        [0.4, 0.2],
        [0.3, 0.4],
        [0.3, 0.6],
        [0.2, 0.5],
        [0.3, 0.6],
        [0.4, 0.5]
    ];
    edges = [
        Path.MOVETO,
        Path.LINETO,
        Path.LINETO,
        Path.LINETO,
        Path.LINETO,
        Path.LINETO,
        Path.LINETO,
        Path.LINETO
    ]
    path = Path(points, edges)
    patch = PathPatch(path, lw=2, edgecolor='black', facecolor='white')
    ax.add_patch(patch)

    c = Circle([0.3, 0.7], 0.075, lw=2, edgecolor='black', facecolor='white')
    ax.add_patch(c)


fig, ax = plt.subplots()
actor(ax, 0.2,0.2)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.grid()
ax.set_xticks( [0.2,0.3,0.4] )
ax.set_yticks( [0.2,0.4,0.6, 0.7, 0.8] )
plt.show()
```
